```python
items = {
         1 : {"Item":"Biscuits", 'Quantity':5,'cost':5},
    
         2 : {"Item":"Cereals", 'Quantity':10,'cost':8},
    
         3 : {"Item":"Chicken", 'Quantity':25,'cost':15},
    
         4 : {"Item":"Oats", 'Quantity':20,'cost':5},
    
         5 : {"Item":"Rice", 'Quantity':30,'cost':6}}
def Customer_details():
    print("Welcome to Dear customer")
    print(" Dear customer,fill your details")
    print("--------------------------------------")
    Name = input("Enter your name: ")
    Address = input(" Enter your adress:")
    Distance = int(input(" Enter your distance from mart(5miles/10miles/15miles/20miles): "))
    return Name,Address,Distance

def items_Available():
    print("Items available in the cart")
    print('----------------------------------------------------------------------------')
    print('%5s %15s %20s %30s' %("S.no","Items","Quantity","Cost for each item"))
    print('----------------------------------------------------------------------------')
    print("%5s %15s %20d %20d"%('1',items[1]['Item'],items[1]['Quantity'],items[1]['cost']))
    print("%5s %15s %20d %20d"%('2',items[2]['Item'],items[2]['Quantity'],items[2]['cost']))
    print("%5s %15s %20d %20d"%('3',items[3]['Item'],items[3]['Quantity'],items[3]['cost']))    
    print("%5s %15s %20d %20d"%('4',items[4]['Item'],items[4]['Quantity'],items[4]['cost']))
    print("%5s %15s %20d %20d"%('5',items[5]['Item'],items[5]['Quantity'],items[5]['cost']))
    return
def Quantities():    
    Biscuits = int(input("\nHow many quantity of Biscuits are  required?: "))
    Cereals = int(input("\nHow many quantity of Cereals are  required?: "))
    Chicken= int(input("\nHow many quantity of Chicken are  required?: "))
    Oats= int(input("\nHow many quantity of Oats are  required?: "))
    Rice= int(input("\nHow many quantity of Rice are  required?: "))
    a=items[1]['Quantity'] - Biscuits
    b=items[2]['Quantity'] - Cereals
    c=items[3]['Quantity'] - Chicken
    d=items[4]['Quantity'] - Oats
    e=items[5]['Quantity'] - Rice
    
    items[1]['Quantity'] = a
    items[2]['Quantity'] = b
    items[3]['Quantity'] = c
    items[4]['Quantity'] = d
    items[5]['Quantity'] = e
    
    
    #items.update({items:'Quantity'})
  
    print('\n--------------------------------------BILL RECIEPT----------------------------------------')
    print('\nName:',Name)
    print('\nAddress:',Address)
    print('-------------------------------------------------------------------------------------------')
    print('Items added in the cart')
    print('-------------------------------------------------------------------------------------------')
    print('%5s %20s %20s %20s %20s' %('\n Sno','List of items','Quantity','your Cost','Remaining items'))
    print('--------------------------------------------------------------------------------------------------')
    print("%5s %20s %15s %20d %20d"%('1',items[1]['Item'],Biscuits,Biscuits*items[1]['cost'],5-Biscuits))
    print("%5s %20s %15s %20d %20d"%('2',items[2]['Item'],Cereals,Cereals*items[2]['cost'],10-Cereals))
    print("%5s %20s %15s %20d %20d"%('3',items[3]['Item'],Chicken,Chicken*items[3]['cost'],25-Chicken))    
    print("%5s %20s %15s %20d %20d"%('4',items[4]['Item'],Oats,Oats*items[4]['cost'],20-Oats))
    print("%5s %20s %15s %20d %20d"%('5',items[5]['Item'],Rice,Rice*items[5]['cost'],30-Rice))
    total_cost =  Biscuits*items[1]['cost'] + Cereals*items[2]['cost'] + Chicken*items[3]['cost'] + Oats*items[4]['cost'] + Rice*items[5]['cost']
    print('--------------------------------------------------------------------------------------------------')
    print()
    print("\n Total Bill Amount")
    return total_cost
def delivery(Distance,total_cost):
    if 0<Distance<=5:
        delivery_charge = 10
        final_cost = total_cost + delivery_charge
       
    elif 5<Distance<=10:
        delivery_charge = 12
        final_cost = total_cost + delivery_charge
        
    elif 10<Distance<=15:
        delivery_charge = 14
        final_cost = total_cost + delivery_charge
        
    else:
        delivery_charge = 20
        final_cost = total_cost + delivery_charge
    print('--------------------------------------------------------------------------------------------------')
    print('%20s %20s %20s ' %('Total cost','Delivery charge','Final cost'))
    print('--------------------------------------------------------------------------------------------------')
    print('%20d %20d %20d ' %(total_cost,delivery_charge,final_cost))
    
    return final_cost
while True:
    Name,Address,Distance=Customer_details()
    items_Available()
    total_cost = Quantities()
    delivery(Distance,total_cost)
    visit_agian = input(" Would you like to do shop again? (yes/no) ")
    if visit_agian == 'no':
        print("---Thanks for your shopping---","\n","***visit again***","\n","---Have a nice day---")
        break
    else:
        continue
```
